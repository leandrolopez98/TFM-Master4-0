```python
"Mover labels de una carpeta a otra"
import os
import shutil

# Directorios de imágenes y etiquetas
train_image_dir = 'Train Images'
train_label_dir = 'Train Labels'
val_image_dir = 'Val Images'
val_label_dir = 'Val Labels'

# Mover las etiquetas de entrenamiento a la carpeta de imágenes de entrenamiento
for label_file in os.listdir(train_label_dir):
    if label_file.endswith('.txt'):
        src = os.path.join(train_label_dir, label_file)
        dst = os.path.join(train_image_dir, label_file)
        shutil.move(src, dst)

# Mover las etiquetas de validación a la carpeta de imágenes de validación
for label_file in os.listdir(val_label_dir):
    if label_file.endswith('.txt'):
        src = os.path.join(val_label_dir, label_file)
        dst = os.path.join(val_image_dir, label_file)
        shutil.move(src, dst)

print("Las etiquetas han sido movidas a las carpetas correspondientes de imágenes.")
```

```python
"Renombrar las clases"
import os

# Directorios de etiquetas
label_dirs = ['Train Labels', 'Val Labels']

# Mapeo de clases según las reglas especificadas
def map_class(class_id):
    if 0 <= class_id <= 16:
        return 0
    elif 17 <= class_id <= 19:
        return 1
    elif class_id == 20:
        return 2
    elif class_id == 21:
        return 3
    elif class_id in [22, 23]:
        return 4
    elif 24 <= class_id <= 27:
        return 5
    else:
        return None  # Si hay alguna clase fuera de este rango, la ignoramos

# Recorrer y modificar todos los archivos de etiquetas en los directorios de entrenamiento y validación
for label_dir in label_dirs:
    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            label_path = os.path.join(label_dir, label_file)
            with open(label_path, 'r') as file:
                lines = file.readlines()
            
            # Modificar las clases según el mapeo
            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) > 0:
                    class_id = int(parts[0])
                    new_class_id = map_class(class_id)
                    if new_class_id is not None:
                        parts[0] = str(new_class_id)
                        new_lines.append(" ".join(parts) + '\n')

            # Guardar los cambios en el archivo de etiquetas
            with open(label_path, 'w') as file:
                file.writelines(new_lines)

print("Las clases han sido renombradas correctamente en todos los archivos de entrenamiento y validación.")
```

```python
"Verificar imagnes con labels"
import os

# Directorios de imágenes y etiquetas
image_dir = 'Train Images'
label_dir = 'Train Labels'

# Verificar que cada imagen tenga un archivo de etiqueta correspondiente
image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.jpg') or f.endswith('.png')])
missing_labels = [img for img in image_files if not os.path.exists(os.path.join(label_dir, os.path.splitext(img)[0] + '.txt'))]

if missing_labels:
    print("Las siguientes imágenes no tienen un archivo de etiqueta correspondiente:")
    print(missing_labels)
else:
    print("Todas las imágenes tienen etiquetas correspondientes.")

# Directorios de validación
val_image_dir = 'Val Images'
val_label_dir = 'Val Labels'

# Verificar que cada imagen de validación tenga un archivo de etiqueta correspondiente
val_image_files = sorted([f for f in os.listdir(val_image_dir) if f.endswith('.jpg') or f.endswith('.png')])
val_missing_labels = [img for img in val_image_files if not os.path.exists(os.path.join(val_label_dir, os.path.splitext(img)[0] + '.txt'))]

if val_missing_labels:
    print("Las siguientes imágenes de validación no tienen un archivo de etiqueta correspondiente:")
    print(val_missing_labels)
else:
    print("Todas las imágenes de validación tienen etiquetas correspondientes.")

empty_labels = []

for label_file in os.listdir(label_dir):
    if label_file.endswith('.txt'):
        label_path = os.path.join(label_dir, label_file)
        with open(label_path, 'r') as file:
            lines = file.readlines()
            if len(lines) == 0:
                empty_labels.append(label_file)

if empty_labels:
    print("Los siguientes archivos de etiquetas están vacíos:")
    print(empty_labels)
else:
    print("No hay archivos de etiquetas vacíos en Train Labels.")
```
