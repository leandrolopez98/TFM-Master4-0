```python
import os

# Directorios de imgenes y etiquetas dentro de Codigo_TFM
image_dirs = ['Train Images', 'Val Images']
label_dirs = ['Train Labels', 'Val Labels']

# Renombrar las etiquetas para que coincidan con las imgenes
for img_dir, lbl_dir in zip(image_dirs, label_dirs):
    # Obtener los nombres de las imgenes y etiquetas
    image_files = sorted([f for f in os.listdir(img_dir) if f.endswith('.jpg') or f.endswith('.png')])
    label_files = sorted([f for f in os.listdir(lbl_dir) if f.endswith('.txt')])
    
    # Asegurar que el nmero de etiquetas y de imgenes sea el mismo
    if len(image_files) != len(label_files):
        print(f"Advertencia: nmero diferente de imgenes y etiquetas en {img_dir}")
    
    # Renombrar cada archivo de etiqueta
    for img_file, lbl_file in zip(image_files, label_files):
        img_name = os.path.splitext(img_file)[0]  # Nombre de la imagen sin extensin
        lbl_name = os.path.splitext(lbl_file)[0]  # Nombre de la etiqueta sin extensin
        new_lbl_path = os.path.join(lbl_dir, f"{img_name}.txt")  # Nuevo nombre para la etiqueta
        
        old_lbl_path = os.path.join(lbl_dir, lbl_file)
        os.rename(old_lbl_path, new_lbl_path)
        print(f"Renombrado: {lbl_file} -> {img_name}.txt")
```

```python
import os

# Obtener la ruta absoluta del directorio actual
abs_path = os.path.abspath('.')
print(f"La ruta absoluta es: {abs_path}")

# Crear el archivo YAML de configuracin con rutas absolutas
data_yaml = f"""
train: {abs_path}/Train Images
val: {abs_path}/Val Images

nc: 6  # Nmero de clases
names: ['botellas de plstico', 'botellas de vidrio', 'frascos', 'latas', 'cartn', 'detergentes']
"""

# Guardar el archivo como Deteccin-residuos.yml
with open('Deteccin-residuos.yml', 'w') as f:
    f.write(data_yaml)

print("Archivo Deteccin-residuos.yml guardado con rutas absolutas.")

```
