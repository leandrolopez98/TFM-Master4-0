```python
# Celda 2c: Creación del Archivo de Configuración (JSON en Formato COCO)
import os
import json
from PIL import Image

def generate_coco_json(images_dir, labels_dir, output_json_path, categories):
    images = []
    annotations = []
    annotation_id = 1

    for image_filename in os.listdir(images_dir):
        if not image_filename.endswith(('.jpg', '.png', '.jpeg')):
            continue

        # Cargar la imagen para obtener su tamaño
        image_path = os.path.join(images_dir, image_filename)
        image = Image.open(image_path)
        width, height = image.size

        # ID de la imagen (extraer desde el nombre del archivo)
        image_id = int(os.path.splitext(image_filename)[0].split('(')[-1].split(')')[0])

        # Añadir información de la imagen
        images.append({
            "id": image_id,
            "file_name": image_filename,
            "width": width,
            "height": height
        })

        # Cargar el archivo de etiqueta correspondiente
        label_filename = os.path.splitext(image_filename)[0] + '.txt'
        label_path = os.path.join(labels_dir, label_filename)

        if os.path.exists(label_path):
            with open(label_path, 'r') as label_file:
                for line in label_file.readlines():
                    class_id, x_center, y_center, bbox_width, bbox_height = map(float, line.split())
                    
                    # Convertir coordenadas relativas a absolutas
                    x_center *= width
                    y_center *= height
                    bbox_width *= width
                    bbox_height *= height
                    
                    # Calcular las coordenadas del borde superior izquierdo
                    x_min = x_center - (bbox_width / 2)
                    y_min = y_center - (bbox_height / 2)

                    # Añadir la anotación
                    annotations.append({
                        "id": annotation_id,
                        "image_id": image_id,
                        "category_id": int(class_id) + 1,  # Ajuste para categorías 1-indexadas
                        "bbox": [x_min, y_min, bbox_width, bbox_height],
                        "area": bbox_width * bbox_height,
                        "iscrowd": 0
                    })
                    annotation_id += 1

    # Crear la estructura del archivo COCO JSON
    coco_json = {
        "images": images,
        "annotations": annotations,
        "categories": categories
    }

    # Guardar el JSON en un archivo
    with open(output_json_path, 'w') as json_file:
        json.dump(coco_json, json_file, indent=4)

# Usar la función con las rutas y categorías
train_images_dir = '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Train-Images/'
train_labels_dir = '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Train-Images/'
val_images_dir = '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Val-Images/'
val_labels_dir = '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Val-Images/'

categories = [
    {"id": 1, "name": "botellas de plástico"},
    {"id": 2, "name": "botellas de vidrio"},
    {"id": 3, "name": "frascos"},
    {"id": 4, "name": "latas"},
    {"id": 5, "name": "cartón"},
    {"id": 6, "name": "detergentes"}
]

# Generar JSON para el conjunto de entrenamiento
generate_coco_json(train_images_dir, train_labels_dir, 'train_annotations_coco.json', categories)

# Generar JSON para el conjunto de validación
generate_coco_json(val_images_dir, val_labels_dir, 'val_annotations_coco.json', categories)
```

```python
# Celda 2c:Dataset Personalizado y DataLoader
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as T

class CustomCocoDataset(Dataset):
    def __init__(self, coco, image_dir, processor, transforms=None):
        self.coco = coco
        self.image_dir = image_dir
        self.processor = processor
        self.transforms = transforms
        self.image_ids = list(coco.imgs.keys())

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        image_id = self.image_ids[idx]
        ann_ids = self.coco.getAnnIds(imgIds=image_id)
        anns = self.coco.loadAnns(ann_ids)
        
        image_info = self.coco.loadImgs(image_id)[0]
        path = image_info['file_name']
        image = Image.open(f"{self.image_dir}/{path}").convert("RGB")
        
        # Obtener las etiquetas y las cajas delimitadoras
        boxes = []
        class_labels = []
        for ann in anns:
            xmin, ymin, width, height = ann['bbox']
            xmax = xmin + width
            ymax = ymin + height
            boxes.append([xmin, ymin, xmax, ymax])
            class_labels.append(ann['category_id'])
        
        # Convertir a tensores
        target = {
            "boxes": torch.tensor(boxes, dtype=torch.float32),
            "class_labels": torch.tensor(class_labels, dtype=torch.int64),
            "image_id": torch.tensor([image_id])
        }

        # Aplicar transformaciones
        if self.transforms is not None:
            image = self.transforms(image)

        return image, target

# Definir transformaciones (si es necesario)
transform = T.Compose([
    T.Resize((800, 800)),
    T.ToTensor(),
])

# Instanciar el dataset para entrenamiento y validación
train_dataset = CustomCocoDataset(train_coco, '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Train-Images', processor, transforms=transform)
val_dataset = CustomCocoDataset(val_coco, '/workspaces/TFM-Master4-0/TFM-Master4-0/Codigo_TFM_2/Val-Images', processor, transforms=transform)

# Crear DataLoaders
train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))
val_dataloader = DataLoader(val_dataset, batch_size=4, shuffle=False, collate_fn=lambda x: tuple(zip(*x)))

```

```python
# Celda 4c: Entrenamiento del Modelo 
import torch
from transformers import DetrForObjectDetection, DetrImageProcessor
from torch.utils.data import DataLoader
from pycocotools.coco import COCO
import torchvision.transforms as T

# Configuración del dispositivo (GPU o CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Cargar el modelo DETR preentrenado y el procesador de imágenes
model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50").to(device)
processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")

# Definir el número de épocas y la tasa de aprendizaje
num_epochs = 10  # Por ejemplo, puedes ajustar este valor según tus necesidades
learning_rate = 5e-5

# Definir el optimizador
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)

# Cargar anotaciones de COCO
train_coco = COCO('train_annotations_coco.json')
val_coco = COCO('val_annotations_coco.json')

# Configuración de entrenamiento
num_epochs = 5  # Puedes ajustar este valor según sea necesario

# ******************************************************************************************
# Reducción del tamaño del lote
# ******************************************************************************************
# NOTA: Hemos reducido el tamaño del lote en los DataLoaders para prevenir problemas de
# sobrecarga de memoria, lo cual podría causar que el kernel se bloquee durante el
# entrenamiento. Si el tamaño del lote es demasiado grande para la memoria disponible
# (RAM o VRAM en la GPU), el kernel puede colapsar, interrumpiendo el proceso de 
# entrenamiento. Este ajuste reduce el tamaño del lote de 4 a 2, pero se puede reducir
# aún más (por ejemplo, a 1) si los problemas persisten. Además, se debe verificar
# que la memoria sea suficiente y optimizar los recursos disponibles.
# ******************************************************************************************

# Crear DataLoaders con un tamaño de lote reducido
train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))
val_dataloader = DataLoader(val_dataset, batch_size=2, shuffle=False, collate_fn=lambda x: tuple(zip(*x)))

# Entrenamiento (Esquema general)
for epoch in range(num_epochs):
    model.train()
    for images, targets in train_dataloader:
        # Procesamiento de las imágenes y etiquetas
        inputs = processor(images, return_tensors="pt").to(device)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        # Adelante
        outputs = model(**inputs, labels=targets)

        # Calcular la pérdida
        loss = outputs.loss

        # Retropropagación
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    print(f"Epoch {epoch + 1} completed. Loss: {loss.item()}")

# Guardar el modelo entrenado
model.save_pretrained("detr_custom_model")
```
